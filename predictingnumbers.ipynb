```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv(r"D:\PythonPrograms\mymachinelearning\Kaggle\digit\numbers\train\train.csv")

```

```python
plt.imshow(data.iloc[115][1:].reshape(28,28))
```

```python
dataX=data.drop('label',axis=1)
```

```python
dataX.head()
```

```python
dataY=data['label']
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
est=SVC()
```

```python
est.fit(X_train,y_train)
```

```python
espredict=est.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
    print(confusion_matrix(y_test,espredict))
    print('\n')
    print(classification_report(y_test,espredict))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc500=RandomForestClassifier(n_estimators=500,verbose=3)
```

```python
rfc500.fit(X_train,y_train)
```

```python
predictRF500=rfc500.predict(X_test)
```

```python
predictRF500
```

```python
print(confusion_matrix(y_test,predictRF500))
print('\n')
print(classification_report(y_test,predictRF500))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, predictRF500, normalize=True, sample_weight=None))
```
